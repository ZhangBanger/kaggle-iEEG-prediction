```python
def gauss_kern_1d(size, factor=1.5):
    """ Returns a normalized 2D gauss kernel array for convolutions """
    x = np.arange(-int(size*factor), int(size*factor)+1)
    g = np.exp(-(x**2/size))
    return g / g.sum()
```

```python
T = 1000
t = np.arange(T)
x = np.random.randn(T)
y = np.convolve(x, gauss_kern_1d(3), mode="same")
```
